```python
# c_text0는 첫번째 기사의 전문
c_text0 = c_texts[0]
print(c_text0)
```

```python
# sents0는 첫번째 기사의 전문을 문장단위로 토큰화
sents0 = nltk.sent_tokenize(c_text0)
sents0
```

```python
import re
```

```python
# 문장 단위로 토큰화 된 것을 regex로 처리해줌(특수문자 제거)
sents0_cleaned = [ re.sub(r"[^a-zA-Z0-9]", " ", sentence) for sentence in sents0 ]
sents0_cleaned
```

```python
#불용어 처리
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python
# sents 는 불용어 처리 이후의 문장들
sents = []
for sentence in sents0_cleaned:
    sentence = " ".join([ word for word in sentence.split() if word.lower() not in stop_words ])
    sents.append(sentence)
sents
```

```python
#토큰화 하기
tokens = []
for sentence in sents:
    tokens.append( nltk.word_tokenize(sentence.lower()) )

tokens
```

```python
# 이중 리스트로 되어 있는 단어들을 1차원 리스트로 변환
all_words = []
for l_tokens in tokens:
    for token in l_tokens:
        all_words.append(token)

all_words
```

```python
#어근 추출
from nltk import PorterStemmer
ps = PorterStemmer()
```

```python
#빈도 확인
frequency = dict()
for word in all_words:
    word = ps.stem(word)
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
frequency
```

```python
#가장 빈도가 높은 값
freq_word = None
most_freq = 0
for key, value in frequency.items():
    if value > most_freq:
        most_freq = value
        freq_word = key
print(most_freq)
print(freq_word)
```

```python
#비율 확인
for key, value in frequency.items():
    frequency[key] = round(value/most_freq, 2)
print(frequency)
```

```python
# 비율이 0.5 이상인것만 저장

imp_words = []
freq_score = 0.5
for key, value in frequency.items():
    if value > freq_score:
        imp_words.append(key)
imp_words
```
