```python
#NLTK를 이용해서 단어만 추출
caesar_words = nltk.corpus.gutenberg.words("shakespeare-caesar.txt")
```

```python
caesar_words
```

```python
type(caesar_words), len(caesar_words)
```

```python
print(caesar_words)
```

```python
#word.isalnum()는 들어있는 값이 문자나 숫자면 돌려주고 아니면 FALSE를 반환한다.
caesar_words = [ word for word in caesar_words if word.isalnum() ]
```

```python
type(caesar_words), len(caesar_words)
```

```python
print(caesar_words)
```

```python
# stopword를 다운받으셔서 위에 처럼 한번 살펴보시기 바랍니다.
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stopwords.fileids()
```

```python
stoplist = stopwords.words("english")
```

```python
type(stoplist), len(stoplist)
```

```python
print(stoplist)
```

```python
#불용어 제거
#word.lower() 소문자 화 시키고 stoplist에 없으면 word에 저장함
# 소문자로 변환한 단어들을 불용어 사전과 비교하여 처리
caesar_words = [ word for word in caesar_words if word.lower() not in stoplist]
```

```python
type(caesar_words), len(caesar_words)
```

```python
#중복된 단어를 제거하고자 함
#set은 중복을 제거해줌
unique_words = set(caesar_words)
```

```python
type(unique_words), len(unique_words)
```

```python
#중복 제거한 unique_words중 50개의 단어만 확인하고 싶음
print(list(unique_words)[:50])
```

```python
# 이해못한 부분
samewords = []
for word in caesar_words:
    for w in unique_words:
        if word == w.lower() and w.isalpha():
            samewords.append(w)
print(len(samewords))
print(samewords)
```

```python
#시저의 단어들을 소문자화해서 caesar_lowercase에 할당한다
caesar_lowercase = set(word.lower() for word in caesar_words)
type(caesar_lowercase), len(caesar_lowercase)
```

```python
print(caesar_lowercase)
```

```python
#이제 단어들이 잘 준비가 되어 있음
#그럼 이 단어들의 분포를 확인한다.(Frequency Distribution)
#FreqDist 클래스는 문서에 사용된 단어(토큰)의 사용빈도 정보를 담는 클래스이다. Text 클래스의 vocab 메서드로 추출할 수 있다.
# FreqDist 이용하여 단어들의 사용빈도 확인
from nltk.probability import FreqDist
freq_caesar = FreqDist(caesar_words)
```

```python
#클래스인걸 보아하니 메서드를 사용할 수 있겠구만
print(type(freq_caesar))
```

```python
#most_common() 함수 - 최빈값 구하기
print(freq_caesar.most_common(100))
```

```python
freq_caesar["Cas"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#가장 빈도가 높은 20개만 top_words에 할당한다
top_words = freq_caesar.most_common(20)
print(type(top_words), len(top_words))
print(top_words)
```

```python
print(list(zip(*top_words)))
```

```python
#단어와 빈도수 분리
#zip() 함수는 여러 개의 순회 가능한(iterable) 객체를 인자로 받고, 각 객체가 담고 있는 원소를 터플의 형태로 차례로 접근할 수 있는 반복자(iterator)를 반환합니다
words = list(zip(*top_words))[0]
frequencies = list(zip(*top_words))[1]
```

```python
print(words)
print(frequencies)
```

```python
x_pos = np.arange(len(words))
x_pos
```

```python
#plt.bar(x, values)
plt.figure(figsize = (40, 15))
#막대 그래프 형태로 x와 y를 지정해준다
#여기에서 x는 x_pos y는 frequencies이다

plt.bar(x_pos, frequencies)
#틱을 설정하는 부분이다. x_pos(즉, x위치에)에 words를 대입해준다(이름설정)
plt.xticks(x_pos, words)
plt.ylabel("Word Count")
plt.xlabel("word")
plt.show()
```
